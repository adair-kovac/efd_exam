```python
import matplotlib.pyplot as plt
import math
import numpy as np

e_0 = 1
c_1 = 2
def tke(time):
	return (c_1*time + 1/math.sqrt(e_0))**-2

times = np.linspace(0, 10, 100000)
tke_at_times = [tke(time) for time in times]

fig, ax = plt.subplots()
plt.loglog(times, tke_at_times)
ax.set_xlabel("log time (s)")
ax.set_ylabel("log tke (m^2/s^2)")
ax.set_title("Turbulent Decay")
```
